```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches

# Load the CSV
csv_path = '/content/drive/MyDrive/IndianPines.csv'  # Update this path
data = pd.read_csv(csv_path)

# Constants
height, width = 145, 145  # Indian Pines dimensions
bands = 200  # spectral bands
label_col = -1  # assuming last column is label

# Extract spectral bands and labels
spectral_data = data.iloc[:, :bands].values
labels = data.iloc[:, label_col].values

# Normalize spectral data for visualization
spectral_data = (spectral_data - spectral_data.min()) / (spectral_data.max() - spectral_data.min())

# Make a false RGB by choosing 3 meaningful bands (e.g., 30, 60, 90)
rgb_indices = [30, 60, 90]
rgb_image = spectral_data[:, rgb_indices].reshape((height, width, 3))

# Ground Truth label map
label_image = labels.reshape((height, width))

# Define color map for 16 classes + background (016)
color_list = [
    '#000080',  # 0 - Background (dark gray)
    '#0000FF',  # 1 - Alfalfa
    '#00BFFF',  # 2 - Corn-notill
    '#008000',  # 3 - Corn-mintill
    '#00FF00',  # 4 - Corn
    '#006400',  # 5 - Grass-pasture
    '#66CDAA',  # 6 - Grass-trees
    '#ADFF2F',  # 7 - Grass-pasture-mowed
    '#FFFF00',  # 8 - Hay-windrowed
    '#FF8C00',  # 9 - Oats
    '#FFA500',  # 10 - Soybean-notill
    '#FFD700',  # 11 - Soybean-mintill
    '#F08080',  # 12 - Soybean-clean
    '#A52A2A',  # 13 - Wheat
    '#FF0000',  # 14 - Woods
    '#8B0000',  # 15 - Buildings-Grass-Trees-Drives
    '#808080'   # 16 - Stone-Steel-Towers
]
cmap = ListedColormap(color_list[:17])  # Class 016

# Class names
class_names = [
    "Background", "Alfalfa", "Corn-notill", "Corn-mintill", "Corn", "Grass-pasture",
    "Grass-trees", "Grass-pasture-mowed", "Hay-windrowed", "Oats", "Soybean-notill",
    "Soybean-mintill", "Soybean-clean", "Wheat", "Woods",
    "Buildings-Grass-Trees-Drives", "Stone-Steel-Towers"
]

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(15, 6), gridspec_kw={'width_ratios': [1, 1, 0.5]})

# RGB Image
axs[0].imshow(rgb_image)
axs[0].set_title('RGB Image')
axs[0].axis('off')

# Ground Truth
im = axs[1].imshow(label_image, cmap=cmap, interpolation='nearest')
axs[1].set_title('Ground Truth')
axs[1].axis('off')

# Legend
legend_patches = [mpatches.Patch(color=color_list[i], label=class_names[i]) for i in range(1, len(class_names))]
axs[2].axis('off')
axs[2].legend(handles=legend_patches, loc='center left', fontsize='large', frameon=False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches

# Load the CSV
csv_path = '/content/drive/MyDrive/Salinas_Dataset.csv'  # Update this path
data = pd.read_csv(csv_path)

# Constants
height, width = 512, 217  # Salinas image dimensions
bands = 204               # Number of spectral bands
label_col = -1            # Assuming last column is label

# Extract spectral bands and labels
spectral_data = data.iloc[:, :bands].values
labels = data.iloc[:, label_col].values

# Normalize for RGB
spectral_data = (spectral_data - spectral_data.min()) / (spectral_data.max() - spectral_data.min())

# Choose RGB bands (e.g., band 30, 60, 90 for visualization)
rgb_indices = [30, 60, 90]
rgb_image = spectral_data[:, rgb_indices].reshape((height, width, 3))

# Ground Truth label map
label_image = labels.reshape((height, width))

# Color list for 16 classes + background (navy blue for background)
color_list = [
    '#000080',  # 0 - Background (navy)
    '#0000FF',  # 1 - Brocoli_green_weeds_1
    '#00BFFF',  # 2 - Brocoli_green_weeds_2
    '#008000',  # 3 - Fallow
    '#00FF00',  # 4 - Fallow_rough_plow
    '#006400',  # 5 - Fallow_smooth
    '#66CDAA',  # 6 - Stubble
    '#ADFF2F',  # 7 - Celery
    '#FFFF00',  # 8 - Grapes_untrained
    '#FF8C00',  # 9 - Soil_vinyard_develop
    '#FFA500',  # 10 - Corn_senesced_green_weeds
    '#FFD700',  # 11 - Lettuce_romaine_4wk
    '#F08080',  # 12 - Lettuce_romaine_5wk
    '#A52A2A',  # 13 - Lettuce_romaine_6wk
    '#FF0000',  # 14 - Lettuce_romaine_7wk
    '#8B0000',  # 15 - Vinyard_untrained
    '#808080'   # 16 - Vinyard_vertical_trellis
]
cmap = ListedColormap(color_list[:17])  # 016

# Class names (match label order)
class_names = [
    "Background", "Brocoli_green_weeds_1", "Brocoli_green_weeds_2", "Fallow", "Fallow_rough_plow",
    "Fallow_smooth", "Stubble", "Celery", "Grapes_untrained", "Soil_vinyard_develop",
    "Corn_senesced_green_weeds", "Lettuce_romaine_4wk", "Lettuce_romaine_5wk",
    "Lettuce_romaine_6wk", "Lettuce_romaine_7wk", "Vinyard_untrained", "Vinyard_vertical_trellis"
]

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(9, 6), gridspec_kw={'width_ratios': [1, 1, 0.5]})

# RGB image
axs[0].imshow(rgb_image)
axs[0].set_title('RGB Image')
axs[0].axis('off')

# Ground Truth
axs[1].imshow(label_image, cmap=cmap, interpolation='nearest')
axs[1].set_title('Ground Truth')
axs[1].axis('off')

# Legend (skip background = index 0)
legend_patches = [mpatches.Patch(color=color_list[i], label=class_names[i]) for i in range(1, len(class_names))]
axs[2].axis('off')
axs[2].legend(handles=legend_patches, loc='center left', fontsize='large', frameon=False)

plt.tight_layout()
plt.show()
```
